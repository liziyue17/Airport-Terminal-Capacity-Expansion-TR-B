```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# population data
df = pd.read_csv('Las Vegas Population.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df['Year'] = df['Date'].dt.year

plt.figure(figsize=(10, 6))

plt.title("Las Vegas Population Data By Year", fontsize=18, fontweight='bold')
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Population", fontsize=14, fontweight='bold')

# Separate the data into two segments
df_real = df[df['Year'] <= 2023]
df_proj = df[df['Year'] >= 2023]


plt.plot(df_real['Year'], df_real['Population'], color='tab:blue', linewidth=1, linestyle='--', marker='o', markersize=4, markerfacecolor='blue', label='Before 2023 (Census Data)')
plt.plot(df_proj['Year'], df_proj['Population'], color='tab:orange', linewidth=1, linestyle='--', marker='s', markersize=4, markerfacecolor='red', label='After 2023 (U.N. Projections)')
# plt.xticks(df['Year'])


years_to_show = list(range(df['Year'].min(), df['Year'].max() + 1, 5))
plt.xticks(years_to_show)

plt.legend()

plt.savefig('Las Vegas Population.pdf', bbox_inches='tight')

plt.show()
```
