```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Annual passenger data
annual_data = pd.read_csv('PHX_Annual.csv')
```

```python
plt.figure(figsize=(10, 6))

plt.title("Passenger Count at PHX", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=10, fontweight='bold')
plt.ylabel("Passenger Count", fontsize=10, fontweight='bold')

plt.ylim(0,50000000)

plt.yticks(np.arange(0, 50000000 + 1, step=5000000))

plt.plot(annual_data['year'], annual_data['total passengers'], label="total passengers", color='tab:orange', linewidth=2.5, linestyle='-', marker='s', markersize=5, markerfacecolor='red')

plt.legend()

plt.savefig('PHX passenger count.pdf', bbox_inches='tight')

plt.show()
```

```python
# Monthly passenger data
monthly_data = pd.read_csv('PHX_Monthly.csv')

monthly_data['date'] = pd.to_datetime(monthly_data[['year', 'month']].assign(day=1))
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(10, 6))

plt.title("Monthly Passenger Data at PHX", fontsize=14, fontweight='bold')
plt.xlabel("Date", fontsize=10, fontweight='bold')
plt.ylabel("Passenger Count", fontsize=10, fontweight='bold')

# plt.ylim(0,50000000)

# plt.yticks(np.arange(0, 50000000 + 1, step=5000000))


plt.plot(monthly_data['date'], monthly_data['enplaned connect'], label='enplaned connect', color='tab:blue', linewidth=1.5, linestyle='-.')
plt.plot(monthly_data['date'], monthly_data['enplaned domestic'], label='enplaned domestic', color='tab:orange', linewidth=1.5, linestyle='-.')
plt.plot(monthly_data['date'], monthly_data['enplaned international'], label='enplaned international', color='tab:cyan', linewidth=1.5, linestyle='-.')
plt.plot(monthly_data['date'], monthly_data['deplaned domestic'], label='deplaned domestic', color='tab:green', linewidth=1.5, linestyle='-.')
plt.plot(monthly_data['date'], monthly_data['deplaned international'], label='deplaned international', color='tab:purple', linewidth=1.5, linestyle='-.')

plt.plot(monthly_data['date'], monthly_data['total'], label="total passengers", color='tab:red', linewidth=1.5, linestyle='-')

# Set x-axis ticks to display every 6 months
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=6))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, fontsize=8, ncol=3)

plt.savefig('PHX monthly passenger count.pdf', bbox_inches='tight')

plt.show()
```
